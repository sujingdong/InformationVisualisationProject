```python
# remove rows with empty abb values, remove rows for Basketball event because its mostly team based and contribute to missing value in gender
df = df[df['abb'].notnull()]
df = df[df['type'] != 'Basketball']

# there were '-' value in abb column which doesnt represent the abbreviation for country
df = df[df['abb'] != '-']

# remove rows with mixed gender type, so we only focus on men and women
df = df[df['gender'] != 'Mixed']

# drop columns that wont be needed for constructing the visualisation
df = df.drop(['athlete', 'grp_id', 'guide', 'country'], axis=1)

# previous step results in duplcated rows when athlete is removed, because for group event competition, the remaining column values are the same because these athletes were in the same team, so I only want to count one medal per group for group event
df = df.drop_duplicates()
```
